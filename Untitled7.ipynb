```python
import os
import shutil
import random

# Define paths
image_dir = "/content/dataset/datasets/images"
train_dir = "/content/dataset/images/train"
val_dir = "/content/dataset/images/val"

# Create train and val directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Get list of all images
all_images = [f for f in os.listdir(image_dir) if f.endswith((".jpg", ".png"))]
random.shuffle(all_images)

# Split into 80% train and 20% val
split_idx = int(0.8 * len(all_images))
train_images = all_images[:split_idx]
val_images = all_images[split_idx:]

# Move images
for img in train_images:
    shutil.move(os.path.join(image_dir, img), os.path.join(train_dir, img))
for img in val_images:
    shutil.move(os.path.join(image_dir, img), os.path.join(val_dir, img))

print(f"Moved {len(train_images)} images to train and {len(val_images)} images to val.")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths (update based on actual structure)
images_train_dir = "/content/dataset/images/train"
images_test_dir = "/content/dataset/images/val"
labels_dir = "/content/dataset/labels"  # Single labels folder (will be split)

train_labels_dir = "/content/dataset/labels/train"
val_labels_dir = "/content/dataset/labels/val"
train_images_dir = "/content/dataset/images/train_split"
val_images_dir = "/content/dataset/images/val_split"

# Create new train/val directories
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)

# Collect all images from both train and test folders
image_files = []
for folder in [images_train_dir, images_test_dir]:
    image_files += [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".jpg")]

# Split into train (80%) and validation (20%)
train_images, val_images = train_test_split(image_files, test_size=0.2, random_state=42)

# Move images and corresponding labels
for img_path in train_images:
    img_name = os.path.basename(img_path)
    label_file = img_name.replace(".jpg", ".txt")

    shutil.move(img_path, os.path.join(train_images_dir, img_name))

    if os.path.exists(os.path.join(labels_dir, label_file)):
        shutil.move(os.path.join(labels_dir, label_file), os.path.join(train_labels_dir, label_file))

for img_path in val_images:
    img_name = os.path.basename(img_path)
    label_file = img_name.replace(".jpg", ".txt")

    shutil.move(img_path, os.path.join(val_images_dir, img_name))

    if os.path.exists(os.path.join(labels_dir, label_file)):
        shutil.move(os.path.join(labels_dir, label_file), os.path.join(val_labels_dir, label_file))

print("âœ… Labels and images successfully split into train and val folders!")
```
